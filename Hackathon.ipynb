### Linear Regression

```python
#Linear Regression

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Assign the features and target variables
X = Clean_Data[['i AGE',
 'i Male',
 'i White',
 'i Education',
 'i Transportation',
 'i No Car',
 'i No Tech',
 'i Marriage']]
y = Clean_Data["i ZEV Sale"]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)

# Create and fit the linear regression model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Print the coefficients
print(lr.coef_)

# Predict the values for the test set
y_pred = lr.predict(X_test)

# Evaluate the model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate the evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test,y_pred)

# Print the results
print("MAE:", mae)
print("MSE:", mse)
print("R-squared:", r2)


```

# Rescale Data

On this part we try to rescale the data between the -1 to 1 instead of 0 to 100 and put out more outline points but as the result it's not get better we don't put it in our presentation

```python
def rescale(df):
    upper_threshold = np.percentile(df, 99)
    lower_threshold = np.percentile(df, 1)
    df[df > upper_threshold] = upper_threshold
    df[df < lower_threshold] = lower_threshold

    return (df - df.min())/(df.max() - df.min())* 2 - 1

Data_rescale = Clean_Data[['i AGE',
 'i Male',
 'i White',
 'i Income',
 'i Education',
 'i Transportation',
 'i No Car',
 'i No Tech',
 'i Marriage',
  "i ZEV Sale" ]].apply(rescale)

Data_rescale_2025 = Data_2025[['i AGE',
 'i Male',
 'i White',
 'i Income',
 'i Education',
 'i Transportation',
 'i No Car',
 'i No Tech',
 'i Marriage']].apply(rescale)
```

### Neural Network on Rescaled Data

```python
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np
from keras.models import Sequential
from keras.layers import Dense

# Assign the features and target variables
X = Data_rescale[[ 'i AGE', 'i Male', 'i White', 'i Income', 'i Education','i Transportation', 'i No Car', 'i No Tech', 'i Marriage']]
y = Data_rescale["i ZEV Sale"]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)

# Initialize the Neural Network model
model = Sequential()

# Add the first layer with 128 neurons and input shape
model.add(Dense(128, input_shape=(X_train.shape[1],), activation='relu'))

# Add the second layer with 128 neurons
model.add(Dense(128, activation='relu'))

# Add the output layer with one neuron
model.add(Dense(1, activation='linear'))

# Compile the model with mean squared error loss function
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model with X_train and y_train
model.fit(X_train, y_train, epochs=100, batch_size=4, verbose=0)

# Predict y_test using the model
y_pred = model.predict(X_test)

# Evaluate the model with mean absolute error and R2 score
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error:", mae)
print("R2 Score:", r2)
```
